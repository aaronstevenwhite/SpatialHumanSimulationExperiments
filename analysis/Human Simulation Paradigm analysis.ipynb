```python
data_humsim = pd.read_csv('../data/norming/human_simulation/results.filtered', sep=';')
data_simil = pd.read_csv('../data/norming/similarity/results.filtered', sep=';')

data_humsim.head()
```

```python
data_simil.head()
```

```python
data_freq = [l.strip().split('\t') for l in open('norming/human_simulation/external_data/sorted.uk.word.unigrams')]

vs = list(data_humsim.verbinflected.unique())

data_freq = pd.DataFrame([w for w in data_freq if w[-1] in vs], columns = ['freq', 'verbinflected'])

data_freq.freq = data_freq.freq.astype(int)

data_freq
```

```python
def zscore(ratings):
    ratings_shifted = ratings - ratings.mean()
    return ratings_shifted / ratings_shifted.std()

data_simil['rating_zscored'] = data_simil.groupby('id').rating.apply(zscore)
```

```python
_ = plt.scatter(data_simil.rating, data_simil.rating_zscored)
```

```python
_ = plt.hist(data_simil_mean.rating_zscored, bins=20)
```

```python
data = pd.merge(data_humsim, data_simil_mean, how='outer', left_on=['verb', 'responselemma'], right_on=['verb0', 'verb1'])

data['nonverb'] = np.logical_and(data.rating_zscored.map(lambda x: np.isnan(x)), 
                                 data.responselemma!=data.verb)

data = pd.merge(data, data_freq)

%Rpush data

data.head()
```

```python
data_learning = pd.read_csv('../data/learning/results.filtered', sep=';')

data_learning['rating_zscored'] = data_learning.groupby('idunique').rating.apply(zscore)

data_learning.head()
```

```python
data_truesim = pd.read_csv('../../ProjectionExperiments/data/likert/likert.filtered')

data_truesim['response_zscored'] = data_truesim.groupby('subj').response.apply(zscore)
data_truesim = pd.concat([data_truesim, data_truesim.rename(columns={'verb0': 'verb1', 'verb1': 'verb0'})])
data_truesim = data_truesim.pivot_table(index=['verb0', 'verb1'], values='response_zscored', aggfunc=np.mean).reset_index()

data_truesim.head()
```

```python
data_learning_nonce = data_learning[data_learning.comptype=='nonce']
data_learning_nonce_truesim = pd.merge(data_learning_nonce, data_truesim, 
                                       left_on=['verb', 'verb1'], right_on=['verb0', 'verb1'])

data_learning_corr = []
head = ['idunique', 'verb', 'context', 'lexical', 'informativity', 'trainingsize']

for name, group in data_learning_nonce_truesim.groupby(head):
    name = list(name)
    name = [name[0][:12]] + name
    stat = sp.stats.spearmanr(group.rating, group.response_zscored)[0]
    stat = stat if not np.isnan(stat) else 0.
    data_learning_corr.append(name + [stat])
    
data_learning_corr = pd.DataFrame(data_learning_corr, columns=['subj']+head+['corr'])

data_learning_corr_mean = data_learning_corr.pivot_table(index=head[2:], values='corr', aggfunc=np.mean).reset_index()

data_learning_corr.head()
```

```python
data_learning_true = data_learning[data_learning.comptype=='True']
data_learning_true_truesim = pd.merge(data_learning_true, data_truesim)

data_learning_corr_true = []

for name, group in data_learning_true_truesim.groupby(head):
    name = list(name)
    name = [name[0][:12]] + name
    stat = sp.stats.spearmanr(group.rating, group.response_zscored)[0]
    stat = stat if not np.isnan(stat) else 0.
    data_learning_corr_true.append(name + [stat])
    
data_learning_corr_true = pd.DataFrame(data_learning_corr_true, columns=['subj']+head+['truecorr'])

data_learning_corr_true.head()
```

```python
data_learning_corr = pd.merge(data_learning_corr, data_learning_corr_true)

%Rpush data_learning_corr data_learning_corr_mean
```
